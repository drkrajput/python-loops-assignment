```python
marks = [50,45,68]
sum(marks)
```

```python
sum_var = 0
n = len(marks)
for i in range(n):
  sum_var += marks[i]

print(sum_var)
```

```python
x = 'Pulkit' # global variable
def full_name(first_name, last_name):
  x = first_name + ' ' + last_name # local variable
  return x

y = full_name('Pulkit', 'Aneja') # first

print(x) # second
print(y)
```

```python
greet_cust(y,'76')
```
