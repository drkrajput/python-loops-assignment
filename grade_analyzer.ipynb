```python
# grade_analyzer.py

def process_scores(students):
    """
    Task 1 — Process the Scores
    Accepts a dictionary {name: [scores]} and returns {name: average_score}.
    """
    averages = {}
    for name, scores in students.items():
        avg = round(sum(scores) / len(scores), 2)
        averages[name] = avg
    return averages


def classify_grades(averages):
    """
    Task 2 — Classify the Grades
    Accepts {name: average_score} and returns {name: (average, grade)}.
    """
    classified = {}

    # Define thresholds inside the function
    grade_A = 90
    grade_B = 75
    grade_C = 60

    for name, avg in averages.items():
        if avg >= grade_A:
            grade = "A"
        elif avg >= grade_B:
            grade = "B"
        elif avg >= grade_C:
            grade = "C"
        else:
            grade = "F"
        classified[name] = (avg, grade)
    return classified


def generate_report(classified, passing_avg=70):
    """
    Task 3 — Generate the Report
    Prints formatted report and returns number of students who passed.
    """
    print("===== Student Grade Report =====")
    passed_count = 0
    for name, (avg, grade) in classified.items():
        status = "PASS" if avg >= passing_avg else "FAIL"
        if status == "PASS":
            passed_count += 1
        print(f"{name:<10} | Avg: {avg:<6} | Grade: {grade} | Status: {status}")
    print("================================")
    total_students = len(classified)
    failed_count = total_students - passed_count
    print(f"Total Students : {total_students}")
    print(f"Passed         : {passed_count}")
    print(f"Failed         : {failed_count}")
    return passed_count


if __name__ == "__main__":
    # Example data
    students = {
        "Alice": [90, 80, 85, 90],
        "Bob": [60, 65, 70, 55],
        "Clara": [95, 92, 97, 101]
    }

    averages = process_scores(students)
    classified = classify_grades(averages)
    generate_report(classified)
```
